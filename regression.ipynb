## Chargement des données

Chargement des données train au format parquet

```python
import pandas as pd

train = pd.read_parquet('train.parquet')
train.head()
```

Chargement des données test au format parquet

```python
import pandas as pd

test = pd.read_parquet('test.parquet')
test.head()
```

## Traitement des données sur les jeu train et test

On change les colonnes ``tpep_pickup_datetime`` et ``tpep_dropoff_datetime`` en ``pickup_year``, ``pickup_month``,``pickup_day``,``pickup_hour`` et respectivement ``dropoff_year``, ``dropoff_month``,``dropoff_day``,``dropoff_hour``. On change également la colonne ``store_and_fwd_flag`` en variable binaire

```python
# transforme les colonnes tpep_pickup_datetime et tpep_dropoff_datetime en pickup_year, pickup_month, pickup_day, pickup_hour et pareil avec dropoff
train['pickup_year'] = pd.to_datetime(train['tpep_pickup_datetime']).dt.year
train['pickup_month'] = pd.to_datetime(train['tpep_pickup_datetime']).dt.month
train['pickup_day'] = pd.to_datetime(train['tpep_pickup_datetime']).dt.day
train['pickup_hour'] = pd.to_datetime(train['tpep_pickup_datetime']).dt.hour
train['dropoff_year'] = pd.to_datetime(train['tpep_dropoff_datetime']).dt.year
train['dropoff_month'] = pd.to_datetime(train['tpep_dropoff_datetime']).dt.month
train['dropoff_day'] = pd.to_datetime(train['tpep_dropoff_datetime']).dt.day
train['dropoff_hour'] = pd.to_datetime(train['tpep_dropoff_datetime']).dt.hour
train['store_and_fwd_flag'] = train['store_and_fwd_flag'].map({'N': 0, 'Y': 1})

test['pickup_year'] = pd.to_datetime(test['tpep_pickup_datetime']).dt.year
test['pickup_month'] = pd.to_datetime(test['tpep_pickup_datetime']).dt.month
test['pickup_day'] = pd.to_datetime(test['tpep_pickup_datetime']).dt.day
test['pickup_hour'] = pd.to_datetime(test['tpep_pickup_datetime']).dt.hour
test['dropoff_year'] = pd.to_datetime(test['tpep_dropoff_datetime']).dt.year
test['dropoff_month'] = pd.to_datetime(test['tpep_dropoff_datetime']).dt.month
test['dropoff_day'] = pd.to_datetime(test['tpep_dropoff_datetime']).dt.day
test['dropoff_hour'] = pd.to_datetime(test['tpep_dropoff_datetime']).dt.hour
test['store_and_fwd_flag'] = test['store_and_fwd_flag'].map({'N': 0, 'Y': 1})
```

On supprime les colonnes ``tpep_pickup_datetime`` et ``tpep_dropoff_datetime`` dont les informations sont maintenant dans les colonnes créées à l'étape précédente

```python
train.drop(['tpep_pickup_datetime','tpep_dropoff_datetime'],axis=1,inplace=True)
test.drop(['tpep_pickup_datetime','tpep_dropoff_datetime'],axis=1,inplace=True)
```

Les données sont déjà néttoyées et ne dispose pas de valeurs manquantes, on peut donc continuer.

## Entrainement
La variable Y contient notre cible ici le pourboire qui est la colonne ``tip_amount`` et X contient nos features. On dispose de 100 000 courses pour l'entrainement (n=100000) et de 25 features (p=25)

```python
Y=train["tip_amount"]
X=train.drop("tip_amount",axis=1)
```

```python
# librairies utilisées à importer
import sklearn
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split,cross_val_score,GridSearchCV, RandomizedSearchCV
```

On sépare notre jeu de données train en données destinés à l'entrainement de notre modèle et en données destinés à l'évaluation de notre modèle.

```python
X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.2)
```

### Modèle Random Forest

On commence par tester un modèle Random Forest connu pour éviter le sur apprentissage et étant robuste étant donné qu'une prédiction est une moyenne de plusieurs prédictions d'arbres de décisions.

#### Fixer les hyperparamètres
Pour fixer les hyperparamètres, je choisis d'utiliser ``RandomizedSearch`` plutot que ``GridSearch`` pour ne pas éviter un temps de recherche des hyperparamètres trop long  

```python
param_dist = {
    'n_estimators': [50, 100, 200, 300],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 5],
    'max_features': ['sqrt', 'log2']
}

model = RandomForestRegressor(random_state=42)

random_search = RandomizedSearchCV(model, param_distributions=param_dist, n_iter=5, cv=5, scoring='r2', random_state=42)
random_search.fit(X_train, y_train)

best_params = random_search.best_params_
print(f"Best hyperparameters: {best_params}")

best_rf_model = random_search.best_estimator_

best_rf_model.fit(X_train, y_train)

y_pred = best_rf_model.predict(X_test)

r2_train = best_rf_model.score(X_train, y_train)
r2_test = r2_score(y_test, y_pred)
cross_val = cross_val_score(best_rf_model, X_train, y_train, cv=5).mean()

print(f'R² on Training Set: {r2_train:.2f}')
print(f'R² on Test Set: {r2_test:.2f}')
print('Cross-validation score:', cross_val)
```

#### Submission Random Forest
On calcule les prédictions de notre modèle Random Forest sur le jeu de données test, on organise pour avoir le format attendu pour le data challenge ici il nous faut une colonne ``row_ID`` qui correspond aux indices des lignes et une colonne ``"tip_amount"``qui correspond à nos prédictions

```python
#soumettre predictions qui contient deux colonnes: row_ID, "tip_amount" au format parquet
predictions=best_rf_model.predict(test)
df=pd.DataFrame(predictions,columns=["\"tip_amount\""])
df.rename(columns={"index": "row_ID"})
df = pd.DataFrame(predictions, columns=["tip_amount"])
df["row_ID"] = df.index
df = df[["row_ID", "tip_amount"]]
df
```

On enregistre nos prédictions dans un format parquet préféré pour ce Data Challenge

```python
df.to_parquet("submission_RandomForest.parquet", index=False)
```

### Modèle XGBoost Regressor



#### Fixer les hyperparamètres

```python
from xgboost import XGBRegressor
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import r2_score

param_dist = {
    'n_estimators': [50, 100, 200, 300],
    'max_depth': [3, 5, 7, 10],
    'learning_rate': [0.01, 0.05, 0.1, 0.2],
    'subsample': [0.6, 0.8, 1.0],
    'colsample_bytree': [0.6, 0.8, 1.0]
}

xgb_model = XGBRegressor(random_state=42)

random_search = RandomizedSearchCV(
    xgb_model, param_distributions=param_dist, n_iter=5, cv=5,
    scoring='r2', random_state=42
)

random_search.fit(X_train, y_train)

best_xgb_model = random_search.best_estimator_
best_params = random_search.best_params_
print(f"Best hyperparameters: {best_params}")

y_pred = best_xgb_model.predict(X_test)

r2_train = best_xgb_model.score(X_train, y_train)
r2_test = r2_score(y_test, y_pred)
cross_val = cross_val_score(best_xgb_model, X_train, y_train, cv=5, scoring='r2').mean()

print(f'R² on Training Set: {r2_train:.2f}')
print(f'R² on Test Set: {r2_test:.2f}')
print('Cross-validation R² score:', cross_val)
```

#### Submission XGBoost Regressor

```python
#soumettre predictions qui contient deux colonnes: row_ID, "tip_amount" au format parquet
import numpy as np
predictions=best_xgb_model.predict(test)
df=pd.DataFrame(predictions,columns=["\"tip_amount\""])
df.rename(columns={"index": "row_ID"})
df = pd.DataFrame(predictions, columns=["tip_amount"])
df["row_ID"] = df.index
df["tip_amount"] = np.where(df["tip_amount"] < 0, 0, df["tip_amount"]) # on vérifie que toutes les valeurs de pourboires sont >=0
df = df[["row_ID", "tip_amount"]]
df
```

```python
df.to_parquet("submission_XGBoost.parquet", index=False)
```

### Modèle LGBM

#### Fixer les hyperparamètres

```python
from lightgbm import LGBMRegressor
from sklearn.model_selection import RandomizedSearchCV, cross_val_score
from sklearn.metrics import r2_score
import pandas as pd

# Définition de l’espace de recherche réduit pour les hyperparamètres
param_dist = {
    'n_estimators': [50, 100, 200, 300],      # Nombre d'arbres
    'max_depth': [None, 10, 20, 30],          # Profondeur maximale des arbres
    'learning_rate': [0.01, 0.05, 0.1, 0.2],  # Taux d'apprentissage
    'subsample': [0.6, 0.8, 1.0],             # Sous-échantillonnage des données
    'colsample_bytree': [0.6, 0.8, 1.0]       # Sous-échantillonnage des colonnes
}

# Initialisation du modèle LGBMRegressor
lgbm_model = LGBMRegressor(random_state=42)

# Recherche aléatoire avec validation croisée
random_search = RandomizedSearchCV(
    lgbm_model, param_distributions=param_dist, n_iter=5, cv=5,  # Nombre d'itérations pour la recherche
    scoring='r2', random_state=42, n_jobs=-1  # Parallélisation pour plus de rapidité
)

# Entraînement avec recherche d’hyperparamètres sur un échantillon (si X_train est trop grand)
random_search.fit(X_train.sample(frac=0.5, random_state=42), y_train.sample(frac=0.5, random_state=42))

# Récupération du meilleur modèle et de ses paramètres
best_lgbm_model = random_search.best_estimator_
best_params = random_search.best_params_
print(f"Best hyperparameters: {best_params}")

# Entraînement final du meilleur modèle sur tout le jeu de données
best_lgbm_model.fit(X_train, y_train)

# Évaluation du modèle avec les données de test
y_pred = best_lgbm_model.predict(X_test)

# Calcul des métriques de performance
r2_train = best_lgbm_model.score(X_train, y_train)
r2_test = r2_score(y_test, y_pred)
cross_val = cross_val_score(best_lgbm_model, X_train, y_train, cv=5, scoring='r2', n_jobs=-1).mean()

print(f'R² on Training Set: {r2_train:.2f}')
print(f'R² on Test Set: {r2_test:.2f}')
print('Cross-validation R² score:', cross_val)
```

#### Submission SVR

```python
#soumettre predictions qui contient deux colonnes: row_ID, "tip_amount" au format parquet
predictions=best_lgbm_model.predict(test)
df=pd.DataFrame(predictions,columns=["\"tip_amount\""])
df.rename(columns={"index": "row_ID"})
df = pd.DataFrame(predictions, columns=["tip_amount"])
df["row_ID"] = df.index
df = df[["row_ID", "tip_amount"]]
df
```

```python
df.to_parquet("submission_LGBM.parquet", index=False)
```

## Agrégation de modèles

Les différents modèles : RandomForest, SVR et XGBoost fournissent différentes prédictions. On va maintenant essayer de moyenner ces modèles

```python
predictions_LGBM=best_lgbm_model.predict(test)
predictions_RF=best_rf_model.predict(test)
predictions_boost=best_xgb_model.predict(test)

predictions_RF[predictions_RF<0]=0
predictions_LGBM[predictions_LGBM<0]=0
predictions_boost[predictions_boost<0]=0

predictions_RF_LGBM=(predictions_LGBM+predictions_RF)/2
predictions_RF_boost=(predictions_boost+predictions_RF)/2
predictions_LGBM_boost=(predictions_boost+predictions_LGBM)/2
predictions_ALL=(predictions_boost+predictions_RF+predictions_LGBM)/3

def submission(predictions,name):
    df=pd.DataFrame(predictions,columns=["\"tip_amount\""])
    df.rename(columns={"index": "row_ID"})
    df = pd.DataFrame(predictions, columns=["tip_amount"])
    df["row_ID"] = df.index
    df = df[["row_ID", "tip_amount"]]
    df.to_parquet(name, index=False)

submission(predictions_RF_LGBM,"submission_RF_LGBM.parquet")
submission(predictions_RF_boost,"submission_RF_boost.parquet")
submission(predictions_LGBM_boost,"submission_LGBM_boost.parquet")
submission(predictions_ALL,"submission_ALL.parquet")
```

```python
pred_lgbm = best_lgbm_model.predict(X_test)
pred_rf=best_rf_model.predict(X_test)
pred_boost=best_xgb_model.predict(X_test)

pred_lgbm[pred_lgbm<0]=0
pred_rf[pred_rf<0]=0
pred_boost[pred_boost<0]=0

pred_lgbm_rf=(pred_lgbm+pred_rf)/2
pred_lgbm_boost=(pred_boost+pred_lgbm)/2
pred_rf_boost=(pred_boost+pred_rf)/2
pred_ALL=(pred_boost+pred_rf+pred_lgbm)/3

r2_test_lgbm_rf = r2_score(y_test, pred_lgbm_rf)
r2_test_lgbm_boost = r2_score(y_test, pred_lgbm_boost)
r2_test_rf_boost = r2_score(y_test, pred_rf_boost)
r2_test_ALL = r2_score(y_test, pred_ALL)

print(f'R² on Test Set LGBM & RF: {r2_test_lgbm_rf:.2f}')
print(f'R² on Test Set LGBM & boost: {r2_test_lgbm_boost:.2f}')
print(f'R² on Test Set RF & boost: {r2_test_rf_boost:.2f}')
print(f'R² on Test Set ALL: {r2_test_ALL:.2f}')
```
